```python
import html

def create_html_table(data):
    # Start the HTML output
    output = "<table>\n"

    # Add the header row
    output += "\t<thead>\n\t\t<tr>\n"
    headers = ["prompt", "index", "content"]
    for header in headers:
        output += "\t\t\t<th>{}</th>\n".format(html.escape(header))
    output += "\t\t</tr>\n\t</thead>\n"

    # Add the data rows
    output += "\t<tbody>\n"
    for row in data:
        output += "\t\t<tr>\n"
        for header in headers:
            output += "\t\t\t<td>{}</td>\n".format(html.escape(str(row[header])))
        output += "\t\t</tr>\n"
    output += "\t</tbody>\n"

    # End the HTML output
    output += "</table>"

    return output
prompt = "How do planes fly?"
# Sample data
data = []
for i in range(len(temp_data)):
    temp_dict = {}
    temp_dict["prompt"] = prompt
    temp_dict["index"] = i + 1
    temp_dict["content"] = temp_data[i]
    data.append(temp_dict)

# Create and print the table
table = create_html_table(data)
print(table)
```
