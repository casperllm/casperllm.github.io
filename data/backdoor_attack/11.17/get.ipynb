```python
flags = [
    "Please",
    "Hello",
    "Thank",
    "There",
    "Greetings",
    "Of course",
    "Sure",
    "Understood",
    "Subject",
    "Dear",
    "Please",
    "Great question!",
]
```

```python
prompts = []
for j in range(len(result["prompts"])):
    contents = result['success_content'][j]
    prompt = result["prompts"][j]
    for i ,content in enumerate(contents):
        if not any([prefix in content for prefix in flags]):
            prompts.append(prompt + " "+ result["success_suffix"][j][i])
```
